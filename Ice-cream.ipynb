```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar dados
dados = pd.read_csv('IceCreamData.csv')
result = dados.sort_values('Temperature', ascending=False)

# Estilo Seaborn
sns.set_style("whitegrid")

# Criar grfico
plt.figure(figsize=(10, 6))

# Plotar linha (agora em vermelho)
plt.plot(result['Temperature'], result['Revenue'], color='red', linewidth=2, label='Vendas')

# Plotar marcadores
plt.scatter(result['Temperature'], result['Revenue'], color='blue', alpha=0.7, label='Pontos de dados')

# Adicionar ttulo e rtulos dos eixos
plt.title('Vendas de Sorvete em Caminhes', fontsize=16)
plt.xlabel('Temperatura (Fahrenheit)', fontsize=14)
plt.ylabel('Vendas (US$)', fontsize=14)

# Adicionar legenda
plt.legend(fontsize=12)

# Adicionar grade
plt.grid(True)

# Ajustar espaamento
plt.tight_layout()

# Exibir o grfico
plt.show()
```
